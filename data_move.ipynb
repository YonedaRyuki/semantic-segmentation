```python
files = glob.glob('/content/drive/MyDrive/3_semantic_segmentation/data/VOCdevkit/VOC2012/21anno_datasets/*')
```

```python
files[0]
```

```python
pil_img = Image.open(files[0])
color = pil_img.getpalette()
img = np.asarray(pil_img)
img_1 = np.where(img != 0,0,img)
img_1 = np.uint8(img_1)
pil_img_color = Image.fromarray(img_1)
pil_img_color = pil_img_color.convert('P')
pil_img_color.putpalette(color)
pil_img_color.save('/content/drive/MyDrive/3_semantic_segmentation/data/VOCdevkit/VOC2012/21anno_datasets_black/'+files[0].split('/')[9])
```

```python
pil_img
```

```python
pil_img_color
```

```python
for i in files:
  pil_img = Image.open(i)
  color = pil_img.getpalette()
  img = np.asarray(pil_img)
  img_1 = np.where(img != 0,0,img)
  img_1 = np.uint8(img_1)
  pil_img_color = Image.fromarray(img_1)
  pil_img_color.convert('P')
  pil_img_color.putpalette(color)
  pil_img_color.save('/content/drive/MyDrive/3_semantic_segmentation/data/VOCdevkit/VOC2012/21anno_datasets_black/'+i.split('/')[9])
```
