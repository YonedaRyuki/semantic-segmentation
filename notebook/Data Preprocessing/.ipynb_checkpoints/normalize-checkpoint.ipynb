```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from scipy.stats import lognorm
from skimage.morphology import remove_small_objects, medial_axis
from astropy.io import fits
from astropy.table import Table, Column
from astropy import units as u
from astropy.wcs import WCS
from astropy.nddata.utils import overlap_slices
from copy import deepcopy
import os
import time
import warnings
from PIL import Image
import glob
```

```python
img = Image.open("semantic_segmentation/data/VOCdevkit/VOC2012/filament_taurus_datasets_grey/taurus_1.jpg")
img_array = np.array(img)


fit_vals = lognorm.fit(img_array[~np.isnan(img_array)])
median = lognorm.median(*fit_vals)
std = lognorm.std(*fit_vals)
thresh_val = median + 2*std

flatten_threshold = thresh_val

flat_array = thresh_val * np.arctan(img_array/flatten_threshold)

flat_norm_array = (flat_array - np.mean(flat_array))/np.std(flat_array)

flat_norm_img = Image.fromarray(flat_norm_array)

flat_norm_img_L = flat_norm_img.convert('L')
```

make dataset normalize  （imsave画像として）うまくいかない

```python
for im in imgs:
    img = Image.open(im)
    img_array = np.array(img)


    fit_vals = lognorm.fit(img_array[~np.isnan(img_array)])
    median = lognorm.median(*fit_vals)
    std = lognorm.std(*fit_vals)
    thresh_val = median + 2*std

    flatten_threshold = thresh_val

    flat_array = thresh_val * np.arctan(img_array/flatten_threshold)

    flat_norm_array = (flat_array - np.mean(flat_array))/np.std(flat_array)
    
    plt.figure()
    
    plt.imshow(flat_norm_array)
    
    plt.imsave("semantic_segmentation/data/VOCdevkit/VOC2012/filament_taurus_datasets_grey_arcnorm/"+im.split("/")[5],flat_norm_array)

```

make dataset normalize  (numpy arrayとして）

```python
for im in imgs:
    img = Image.open(im)
    img_array = np.array(img)


    fit_vals = lognorm.fit(img_array[~np.isnan(img_array)])
    median = lognorm.median(*fit_vals)
    std = lognorm.std(*fit_vals)
    thresh_val = median + 2*std

    flatten_threshold = thresh_val

    flat_array = thresh_val * np.arctan(img_array/flatten_threshold)

    flat_norm_array = (flat_array - np.mean(flat_array))/np.std(flat_array)
    
    np.save("semantic_segmentation/data/VOCdevkit/VOC2012/filament_taurus_datasets_grey_arcnorm_array/"+im.split("/")[5].split(".")[0],flat_norm_array)
```
